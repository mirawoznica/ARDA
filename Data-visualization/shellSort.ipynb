```python
import json
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../JavaExperiment/Data/sorting_times.csv')

df.columns = df.columns.str.strip()
sort_times = df[df['Algorithm'] == 'Shell Sort'].iloc[0, 1:].values

array_lengths = np.arange(1, len(sort_times) + 1)

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(array_lengths, sort_times, marker='o', linestyle='-', color='#1b5886', label='Shell Sort Times')

plt.xticks()
plt.yticks(np.arange(min(sort_times), max(sort_times), 5000))

plt.xlabel('Array Number', fontsize=12)
plt.ylabel('Time (nanoseconds)', fontsize=12)
plt.title('Shell Sort Performance Over Various Array Sizes')
plt.legend()
plt.grid(True)

# Saving the plot as a png file
output_file_path =  '../Data-visualization/Graphs/ShellSortLineGraph.png'
plt.savefig(output_file_path)

plt.show()
```
