```python
import json
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/sorting_algorithm_times.csv')

df.columns = df.columns.str.strip()
bubble_sort_times = df[df['Algorithm'] == 'Insertion Sort'].iloc[0, 1:].values

# Assuming array lengths are equal to the index + some base length
array_lengths = np.arange(1, len(bubble_sort_times) + 1)

# Plotting the line graph for Bubble Sort times
plt.figure(figsize=(10, 6))
plt.plot(array_lengths, bubble_sort_times, marker='o', linestyle='-', color='#1b5886', label='Bubble Sort Times')

plt.xticks(array_lengths)
plt.yticks(np.arange(min(bubble_sort_times), max(bubble_sort_times) + 250, 250))

plt.xlabel('Array Number', fontsize=12)
plt.ylabel('Time (nanoseconds)', fontsize=12)
plt.title('Bubble Sort Performance Over Various Array Sizes')
plt.legend()
plt.grid(True)

# Assuming you want to save the plot as a .png file
output_file_path =  '../Data/BubbleSortLineGraph.png'
plt.savefig(output_file_path)

plt.show()
```
